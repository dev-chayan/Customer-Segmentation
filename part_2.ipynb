## Feature Engineering

### KPI determination

#### 1. Ulilisation_ratio = balance / limit 

utilisation ratio means how much credit one customer is used from hisd limit.

As your credit utilization increases, your credit score can go down. A high credit utilization indicates that you're probably spending a significant portion of your monthly income on debt payments, and this puts you at a higher risk of defaulting on your payments (at least in the eyes of creditors).

```python
data['utilisation_ratio'] = data['BALANCE'] / data['CREDIT_LIMIT']
```

```python
data['utilisation_ratio'].plot(kind = 'hist')
```

```python
data['utilisation_ratio'].plot(kind = 'kde')
```

```python
data[data['utilisation_ratio'] > 1].sort_values(by = 'utilisation_ratio', ascending = False)
```

```python
# Now we decide the cut offs
cutoff_1 = .28
cutoff_2 = .68
cutoff_3 = 1.01 # (we care taking .01 to avoid round off error)
def categorise_utilisation_ratio(cutoff_1,cutoff_2,cutoff_3,data_point):
    if data_point <= cutoff_1:
        return "Low"
    elif data_point <= cutoff_2:
        return "Moderate"
    elif data_point <= cutoff_3:
        return "High"
    else:
        return "Extreme"
```

```python
data['utilisation_ratio_type'] = data['utilisation_ratio'].apply(lambda x: categorise_utilisation_ratio(cutoff_1,cutoff_2,cutoff_3,x))
```

#### 4. Purchase_type 
- To find what type of purchases customers are making on credit card,lets explore the data.

```python
print(data[(data['ONEOFF_PURCHASES']==0) & (data['INSTALLMENTS_PURCHASES']==0)].shape)
print(data[(data['ONEOFF_PURCHASES']>0) & (data['INSTALLMENTS_PURCHASES']>0)].shape)
print(data[(data['ONEOFF_PURCHASES']>0) & (data['INSTALLMENTS_PURCHASES']==0)].shape)
print(data[(data['ONEOFF_PURCHASES']==0) & (data['INSTALLMENTS_PURCHASES']>0)].shape)
```

**I found out that there are 4 types of purchase behaviour in the data set.**
               
               1.People who only do One-Off Purchases.
               2.People who only do Installments Purchases.
               3.People who do both.
               4.People who do none.
               
**So deriving a categorical variable based on the behaviour.**

```python
def purchase(credit):   
    if (credit['ONEOFF_PURCHASES']==0) & (credit['INSTALLMENTS_PURCHASES']==0):
        return 'none'
    if (credit['ONEOFF_PURCHASES']>0) & (credit['INSTALLMENTS_PURCHASES']>0):
         return 'both_oneoff_installment'
    if (credit['ONEOFF_PURCHASES']>0) & (credit['INSTALLMENTS_PURCHASES']==0):
        return 'one_off'
    if (credit['ONEOFF_PURCHASES']==0) & (credit['INSTALLMENTS_PURCHASES']>0):
        return 'istallment'
```

```python
data['purchase_type'] = data.apply(purchase,axis=1)
```

```python
## new dataset considering first 10 principal components
pca_data1 = pd.DataFrame(pca_data[:,:10],columns = [f'PC{i}' for i in range(1,11)])
pca_data1
```

```python
## lets make function for this results
 

def get_summary_kmeans(n_clusters,color_map):
    from sklearn.cluster import KMeans
    km_3=KMeans(n_clusters=n_clusters,random_state=123,init='k-means++',n_init = 15,max_iter = 1000)
    km_3.fit(pca_data1)
    km_3.labels_
    print(pd.Series(km_3.labels_).value_counts())
    # Lets visualize clusters based on 1st 2 principle componenets
    label_color=[color_map[l] for l in km_3.labels_]
    plt.figure(figsize=(7,7))
    plt.scatter(pca_data1['PC1'],pca_data1['PC2'],c=km_3.labels_,cmap='Spectral',alpha=0.5)
    plt.show()
    pca_km3_data = pca_data1.copy()
    pca_km3_data['Cluster']=km_3.labels_
    sns.pairplot(pca_km3_data,hue='Cluster', palette= 'Dark2', diag_kind='kde',size=1.85)
    plt.show()
    ## Now concat cluster labels in main_data
    main_km3_data = data_raw_processed.copy()  
    main_km3_data['cluster'] = km_3.labels_
    print(main_km3_data.groupby('cluster').median().T)
    print("\n\n")
    print(main_km3_data.groupby('cluster').mean().T)
```

#### elbow method

```python
within_sum_squares_list = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k,random_state=123,init='k-means++',n_init = 15,max_iter = 1000)
    kmeans.fit(pca_data1)
    within_sum_squares_list.append(kmeans.inertia_)
```

####  silhouette coefficient

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_scores_list = []
for k in range(2,10):
    kmeans = KMeans(n_clusters=k,random_state=123,init='k-means++',n_init = 15,max_iter = 1000)
    kmeans.fit(pca_data1)
    score = silhouette_score(pca_data1,kmeans.labels_)
    silhouette_scores_list.append(score)
```

```python
pd.Series(silhouette_scores_list).plot(kind = 'line')
```

```python
from sklearn.mixture import GaussianMixture 
```

```python
def SelBest(arr:list, X:int)->list:
    '''
    returns the set of X configurations with shorter distance
    '''
    dx=np.argsort(arr)[:X]
    return arr[dx]
```

```python
from sklearn import metrics
n_clusters=np.arange(2, 10)
sils=[]
sils_err=[]
iterations=20
for n in n_clusters:
    tmp_sil=[]
    for _ in range(iterations):
        gmm=GaussianMixture(n, n_init=10,max_iter  = 500).fit(pca_data1) 
        labels=gmm.predict(pca_data1)
        sil=metrics.silhouette_score(pca_data1, labels, metric='euclidean')
        tmp_sil.append(sil)
    val=np.mean(SelBest(np.array(tmp_sil), int(iterations/5)))
    err=np.std(tmp_sil)
    sils.append(val)
    sils_err.append(err)
```

```python
plt.errorbar(n_clusters, sils, yerr=sils_err)
plt.title("Silhouette Scores", fontsize=20)
plt.xticks(n_clusters)
plt.xlabel("N. of clusters")
plt.ylabel("Score")
```

```python
from sklearn import metrics
n_clusters=np.arange(2, 10)
sils=[]
sils_err=[]
iterations=20
for n in n_clusters:
    tmp_sil=[]
    for _ in range(iterations):
        dbscan =DBSCAN(eps = .5,min_samples= 5).fit(pca_data1) 
        labels=dbscan.labels_
        sil=metrics.silhouette_score(pca_data1, labels, metric='euclidean')
        tmp_sil.append(sil)
    val=np.mean(SelBest(np.array(tmp_sil), int(iterations/5)))
    err=np.std(tmp_sil)
    sils.append(val)
    sils_err.append(err)
```

```python
plt.errorbar(n_clusters, sils, yerr=sils_err)
plt.title("Silhouette Scores", fontsize=20)
plt.xticks(n_clusters)
plt.xlabel("N. of clusters")
plt.ylabel("Score")
```
